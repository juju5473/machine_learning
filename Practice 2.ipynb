```python
# uniform distribution
uniform_values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(uniform_values, 50)
plt.show()
```

```python
# normal distribution
from scipy.stats import norm

x = np.arange(-3, 3, 0.001)   # Returns evenly spaced values within the interval
plt.plot(x,norm.pdf(x))       # pdf is continuous vs. relative frequency histogram
plt.show()

normal_values = np.random.normal(0, 1, 6000)  # for the above pdf: 6 x Std = 3+3 ==> Std=1
plt.hist(normal_values, 50)   # histogram is discrete, dividing the values by the area underneath the histogram returns the relative frequency histogram
plt.show()
```

```python
random_values_array = np.random.normal(100, 20, 100000)       # np.random.normal(mean, std, # of elements)
plt.hist(random_values_array, 50)
plt.show()

print(np.percentile(random_values_array, 2.5))
print(np.percentile(random_values_array, 16))
print(np.percentile(random_values_array, 50))
print(np.percentile(random_values_array, 84)) 
print(np.percentile(random_values_array, 97.5))
```

```python
# linear algebra
first_var = np.array([8,5,4])
sec_var = np.array([0,-2,14])
print(np.dot(first_var,sec_var))
print(np.inner(first_var,sec_var))
print(np.cross(first_var,sec_var))
print(np.dot(first_var,np.cross(first_var,sec_var)))
print(np.dot(sec_var,np.cross(first_var,sec_var)))
```

```python
# covariance and correleation
ind_var = np.random.normal(100, 10, 1000)
dep_var = np.random.normal(80, 10, 1000)*0.15*ind_var + 1.4

print("Vaiance of ind_var is:", ind_var.var(),'\n')  # print with line break
print("Vaiance of dep_var is:", dep_var.var(),'\n')
print("Covaiance cov(ind_var,dep_var) is:", np.cov(ind_var,dep_var)[0][1],'\n')
print("Correlation of ind_var an dep_var is:", np.cov(ind_var,dep_var)[0][1]/(ind_var.std()*dep_var.std()),'\n')

plt.plot(ind_var,dep_var,'.')  
plt.show()
```

```python
from numpy.linalg import inv
from numpy.linalg import det
from sympy import Matrix

a_mat = np.array([[1., 2.], [3., 4.]])
b_mat = np.array([[2, 1, 0], [0, 2, 0], [2, 0, 1]])
```
